```python
# hide
import altair as alt
import pandas as pd
from altair import Chart, X, Y, Axis, SortField
```

```python
# hide
df_bs = pd.read_csv("https://raw.githubusercontent.com/openZH/covid_19/master/fallzahlen_kanton_total_csv/COVID19_Fallzahlen_Kanton_BS_total.csv")
df_bl = pd.read_csv("https://raw.githubusercontent.com/openZH/covid_19/master/fallzahlen_kanton_total_csv/COVID19_Fallzahlen_Kanton_BL_total.csv")
```

```python
# hide
df_bs.head()
```

```python
# hide
df_bl.head()
```

```python
# hide
chart = alt.Chart(pd.concat([df_bs, df_bl])).mark_line(interpolate='step-after').encode(
        x=X('date:T', axis=Axis(title='Date')),
        y=Y('ncumul_conf', axis=Axis(title="Confirmed cases (accumulated)")),
        color=alt.Color('abbreviation_canton_and_fl', legend=alt.Legend(title="Cantons"))
    ).properties(title="Confirmed cases in BS and BL")
```
